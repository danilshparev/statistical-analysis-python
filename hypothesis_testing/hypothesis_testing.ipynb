```python
#Единицы измерения: усл. ед.
#Объём выборки: n = 102
#Уровень значимости: α = 0,05

#Выполнение
#1) По столбцам X1 и X2 проверить гипотезу о равенстве математических ожиданий двух независимых нормальных совокупностей.
#2) По столбцу X3 проверить гипотезу:
   #H₀: sigma^2 = 157,65.
#3) По столбцу X4 проверить гипотезу:
   #H₀: X4 имеет показательное распределение с параметром λ = 0,176.
   #Использовать критерий согласия Пирсона χ².
```

```python
 #4.2: проверка по столбцам Х1 и Х2 гипотезы о равенстве мат ожиданий.

import pandas as pd
import numpy as np
import scipy.stats as stats

df = pd.read_csv('RGR2_D-8_X1-X4.csv', sep=';')
alpha = 0.05
n1 = n2 = 102

print("H0: EX1 = EX2")
print("H1: EX1 != EX2")
print(f"Параметрический критерий: двухвыборочный t-критерий Стьюдента для независимых выборок")
print("   Обоснование: Данные количественные, выборки независимы, предполагается нормальность распределения")


#Проверка гипотезы

x1 = df['X1']
x2 = df['X2']

m1, m2 = x1.mean(), x2.mean()
s1_var, s2_var = x1.var(ddof=1), x2.var(ddof=1)

# Объединенная оценка дисперсии
s_pooled_sq = ((n1 - 1) * s1_var + (n2 - 1) * s2_var) / (n1 + n2 - 2)
# Наблюдаемая t-статистика
t_obs = (m1 - m2) / np.sqrt(s_pooled_sq * (1/n1 + 1/n2))

# Нахождение p-value и критической области
df_degrees = n1 + n2 - 2
p_value = 2 * (1 - stats.t.cdf(abs(t_obs), df_degrees))
t_crit = stats.t.ppf(1 - alpha/2, df_degrees)

print(f"Среднее X1: {m1:.4f}")
print(f"Среднее X2: {m2:.4f}")
print(f"Наблюдаемое значение статистики (t_набл): {t_obs:.4f}")
print(f"Критическое значение (t_крит) для alpha={alpha}: {t_crit:.4f}")
print(f"Критическая область W: (-бесконечность; -{t_crit:.4f}] U [{t_crit:.4f}; +бесконечность)")
print(f"p-value: {p_value:.4f}")
print("Ошибка 1 рода: отклонить гипотезу о равенстве средних, когда на самом деле они равны.")
print("Выводы:")
if abs(t_obs) > t_crit:
    print(f"Так как |t_набл| > t_crit ({abs(t_obs):.4f} > {t_crit:.4f}), H0 отвергается.")
    print("Вывод: Существуют статистически значимые основания считать средние значения X1 и X2 различными.")
else:
    print(f"Так как |t_набл| <= t_crit ({abs(t_obs):.4f} <= {t_crit:.4f}), H0 не отвергается.")
    print(f"Так как p-value ({p_value:.4f}) > alpha ({alpha}), H0 не отвергается.")
    print("Вывод: нет статистически значимых оснований считать средние различными")
```

```python
#4.3
sigma2_0 = 157.65
x3 = df['X3']
n = 102
alpha = 0.05
print(f"H0: sigma^2 = {sigma2_0}")
print(f"H1: sigma^2 != {sigma2_0}")

s2 = x3.var(ddof=1)
print(f"Выборочная дисперсия s^2: {s2:.4f}")

# Статистика критерия
# Статистика хи-квадрат: chi2 = (n-1)*s^2 / sigma2_0
chi2_obs = (n - 1) * s2 / sigma2_0
print(f"Наблюдаемое значение статистики хи^2_набл: {chi2_obs:.4f}")

# Критические значения и p-value
df_degrees = n - 1
chi2_crit_left = stats.chi2.ppf(alpha/2, df_degrees)
chi2_crit_right = stats.chi2.ppf(1 - alpha/2, df_degrees)
p_value = 2 * min(stats.chi2.cdf(chi2_obs, df_degrees), 1 - stats.chi2.cdf(chi2_obs, df_degrees))

print(f"Критический интервал: ({chi2_crit_left:.4f}, {chi2_crit_right:.4f})")
print(f"p-value: {p_value:.4f}")
print("Ошибка 1 рода: отклонить гипотезу о равенстве дисперсии 157.65, когда на самом деле она верна.")
# Выводы
if chi2_obs < chi2_crit_left or chi2_obs > chi2_crit_right:
    print("Результат: Статистика попадает в критическую область. H0 отвергается.")
else:
    print("Результат: Статистика не попадает в критическую область. H0 не отвергается.")
```

```python
#4.4
# Критерий Манна-Уитни
x1 = df['X1']
x2 = df['X2']

# Выполнение проверки
u_stat, p_val_mw = stats.mannwhitneyu(x1, x2, alternative='two-sided')

print(f"Статистика U (Манна-Уитни): {u_stat:.4f}")
print(f"p-value: {p_val_mw:.4f}")
print("Ошибка 1 рода: отклонить гипотезу об одинаковости распределений, когда на самом деле они не различаются.")
if p_val_mw < alpha:
    print("Результат: H0 отвергается. Распределения выборок имеют значимые различия.")
else:
    print("Результат: H0 не отвергается. Значимых различий в распределениях не обнаружено.")

print("\n3. Сравнение с пунктом 4.2:")
print("Выводы в 4.2 (t-тест) и 4.4 (Манна-Уитни) совпали. Это объясняется тем, что")
print("данные специально подготовлены (нормализованы), а критерий Манна-Уитни")
print("обладает высокой мощностью, близкой к t-критерию, на нормальных выборках.")
```

```python
#4.5
#H0: X4 имеет показательное распределение с lambda = 0.176

x4 = df['X4']
n = 102
lambda_hyp = 0.176
alpha = 0.05

print(f"H0: X4 имеет показательное распределение с lambda = {lambda_hyp}")
print(f"H1: X4 не имеет показательного распределения (двусторонняя альтернатива)")
print(f"Критерий: Критерий согласия Пирсона хи-квадрат")
print(f"Обоснование: Проверка соответствия эмпирических частот теоретическому закону распределения")
print("Ошибка 1 рода: отвергнуть гипотезу о показательном распределении, когда на самом деле выборка ему соответствует.\n")
# Определение числа интервалов по правилу Стерджеса
k = int(1 + np.log2(n))

# Разбиение на интервалы: от 0 до максимального значения выборки, последний интервал до бесконечности
# для того, чтобы получился корректный учет хвоста показательного распределения
bins = np.concatenate([np.linspace(0, x4.max(), k), [np.inf]])
counts, _ = np.histogram(x4, bins=bins)

# Теоретические вероятности для интервалов [a, b) (последний интервал [a, +∞))
probs = []
for i in range(len(bins)-1):
    a = bins[i]
    b = bins[i+1]
    if np.isinf(b):
        p = np.exp(-lambda_hyp * a)
    else:
        p = np.exp(-lambda_hyp * a) - np.exp(-lambda_hyp * b)
    probs.append(p)

# Ожидаемые частоты
expected = n * np.array(probs)

# Объединение интервалов с ожидаемыми частотами < 5
merged_counts = []
merged_expected = []
merged_intervals = []

current_count = 0
current_expected = 0.0
left_edge = bins[0]

for i in range(len(counts)):
    current_count += counts[i]
    current_expected += expected[i]
    right_edge = bins[i+1]
    if current_expected >= 5 or i == len(counts)-1:
        if i == len(counts)-1 and current_expected < 5 and len(merged_expected) > 0:
            merged_counts[-1] += current_count
            merged_expected[-1] += current_expected
            merged_intervals[-1] = f"[{merged_intervals[-1].split('[')[1].split(',')[0]}, {right_edge if not np.isinf(right_edge) else '+бесконечность'})"
        else:
            merged_counts.append(current_count)
            merged_expected.append(current_expected)
            merged_intervals.append(f"[{left_edge:.2f}, {right_edge if not np.isinf(right_edge) else '+бесконечность'})")
        current_count = 0
        current_expected = 0.0
        if i+1 < len(bins):
            left_edge = bins[i+1]

# Вычисление статистики хи-квадрат вручную
chi2_obs = sum((np.array(merged_counts) - np.array(merged_expected))**2 / np.array(merged_expected))
df_degrees = len(merged_counts) - 1
chi2_crit = stats.chi2.ppf(1 - alpha, df_degrees)
p_value_pirson = stats.chi2.sf(chi2_obs, df_degrees)

print("Интервалы после объединения (для обеспечения ожидаемой частоты >= 5):")
for iv in merged_intervals:
    print(f"  {iv}")
print(f"Наблюдаемые частоты: {[int(x) for x in merged_counts]}")
print(f"Ожидаемые частоты: {[round(float(e),2) for e in merged_expected]}")
print(f"Наблюдаемое значение статистики chi2_набл: {chi2_obs:.4f}")
print(f"Критическое значение chi2_крит: {chi2_crit:.4f}")
print(f"p-value: {p_value_pirson:.4f}")

print("\nВыводы:")
if chi2_obs > chi2_crit:
    print(f"Так как chi2_набл > chi2_крит ({chi2_obs:.4f} > {chi2_crit:.4f}), H0 отвергается.")
    print(f"Так как p-value ({p_value_pirson:.4f}) < alpha ({alpha}), H0 отвергается.")
    print("Вывод: Выборка X4 не согласуется с показательным распределением (lambda=0.176).")
else:
    print(f"Так как chi2_набл <= chi2_крит ({chi2_obs:.4f} <= {chi2_crit:.4f}), H0 не отвергается.")
    print("Вывод: Нет оснований считать, что распределение X4 отличается от показательного.")
```
